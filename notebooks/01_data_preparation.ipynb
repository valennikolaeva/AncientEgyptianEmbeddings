```python
import os
import json
import pandas as pd
import gc
from pathlib import Path

def extract_flat_passport_info(passport_data):
    """ĞŸĞ¾Ğ»Ğ½Ğ¾ÑÑ‚ÑŒÑ Ñ€Ğ°ÑĞ¿Ğ°ĞºĞ¾Ğ²Ñ‹Ğ²Ğ°ĞµÑ‚ Ğ¿Ğ°ÑĞ¿Ğ¾Ñ€Ñ‚ Ğ² Ğ¿Ğ»Ğ¾ÑĞºÑƒÑ ÑÑ‚Ñ€ÑƒĞºÑ‚ÑƒÑ€Ñƒ"""
    result = {}

    def extract_recursive(children, prefix="passport_"):
        for child in children:
            item_type = child.get('type', '')
            item_value = child.get('value', '')
            item_id = child.get('_id', '')

            if item_type:
                key = f"{prefix}{item_type}"
                if item_value:
                    result[key] = item_value
                if item_id:
                    result[f"{key}_id"] = item_id

            # Ñ€ĞµĞºÑƒÑ€ÑĞ¸Ğ²Ğ½Ğ¾ Ğ¾Ğ±Ñ€Ğ°Ğ±Ğ°Ñ‚Ñ‹Ğ²Ğ°ĞµÑ‚ Ğ²Ğ»Ğ¾Ğ¶ĞµĞ½Ğ½Ñ‹Ğµ ÑÑ‚Ñ€ÑƒĞºÑ‚ÑƒÑ€Ñ‹
            if 'children' in child:
                new_prefix = f"{prefix}{item_type}_" if item_type else prefix
                extract_recursive(child['children'], new_prefix)

    if passport_data and 'children' in passport_data:
        extract_recursive(passport_data['children'])

    return result

def extract_flat_text_content(text_content):
    """ĞŸĞ¾Ğ»Ğ½Ğ¾ÑÑ‚ÑŒÑ Ñ€Ğ°ÑĞ¿Ğ°ĞºĞ¾Ğ²Ñ‹Ğ²Ğ°ĞµÑ‚ Ñ‚ĞµĞºÑÑ‚ Ğ² Ğ¿Ğ»Ğ¾ÑĞºÑƒÑ ÑÑ‚Ñ€ÑƒĞºÑ‚ÑƒÑ€Ñƒ"""
    result = {
        'sentence_count': 0,
        'word_count': 0,
        'all_translations': [],
        'all_hieroglyphic_words': [],
        'sentence_details': []
    }

    for sentence_idx, sentence in enumerate(text_content.get('textItems', [])):
        result['sentence_count'] += 1
        sentence_info = {
            'sentence_id': sentence.get('_id', ''),
            'sentence_type': sentence.get('type', ''),
            'sentence_state': sentence.get('state', ''),
            'translations': [],
            'words': []
        }

        for trans in sentence.get('translation', {}).get('translations', []):
            trans_text = trans.get('value', '')
            if trans_text:
                sentence_info['translations'].append(trans_text)
                result['all_translations'].append(trans_text)

        for item in sentence.get('sentenceItems', []):
            if 'BTSWord' in item.get('eClass', ''):
                word_data = {
                    'word': item.get('wChar', ''),
                    'lKey': item.get('lKey', ''),
                    'flexCode': item.get('flexCode', ''),
                    'graphics': ' '.join(g.get('code', '') for g in item.get('graphics', [])),
                    'word_translations': ' | '.join(t.get('value', '') for t in item.get('translation', {}).get('translations', []))
                }
                sentence_info['words'].append(word_data)
                result['all_hieroglyphic_words'].append(item.get('wChar', ''))
                result['word_count'] += 1

            elif 'BTSMarker' in item.get('eClass', ''):

                marker_data = {
                    'marker_type': item.get('type', ''),
                    'marker_name': item.get('name', '')
                }
                sentence_info.setdefault('markers', []).append(marker_data)

        result['sentence_details'].append(sentence_info)

    result['all_translations_combined'] = ' | '.join(result['all_translations'])
    result['all_hieroglyphic_combined'] = ' | '.join(result['all_hieroglyphic_words'])

    return result

def extract_flat_relations(relations):
    """Ğ Ğ°ÑĞ¿Ğ°ĞºĞ¾Ğ²Ñ‹Ğ²Ğ°ĞµÑ‚ relations Ğ² Ğ¿Ğ»Ğ¾ÑĞºÑƒÑ ÑÑ‚Ñ€ÑƒĞºÑ‚ÑƒÑ€Ñƒ"""
    relations_list = []
    for rel in relations:
        relations_list.append(f"{rel.get('type', '')}:{rel.get('objectId', '')}")
    return ' | '.join(relations_list)

def process_single_file_flat(file_path):
    """ĞĞ±Ñ€Ğ°Ğ±Ğ°Ñ‚Ñ‹Ğ²Ğ°ĞµÑ‚ Ğ¾Ğ´Ğ¸Ğ½ Ñ„Ğ°Ğ¹Ğ» Ñ Ğ¿Ğ¾Ğ»Ğ½Ğ¾Ğ¹ Ñ€Ğ°ÑĞ¿Ğ°ĞºĞ¾Ğ²ĞºĞ¾Ğ¹"""
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)

        records = []

        for text_unit in data:
            record = {
                'file_name': os.path.basename(file_path),
                'file_path': str(file_path),
                'text_unit_id': text_unit.get('_id', ''),
                'name': text_unit.get('name', ''),
                'type': text_unit.get('type', ''),
                'revisionState': text_unit.get('revisionState', ''),
                'state': text_unit.get('state', ''),
                'corpusPrefix': text_unit.get('corpusPrefix', ''),
                'owner': text_unit.get('owner', ''),
                'relations_flat': extract_flat_relations(text_unit.get('relations', [])),
            }

            passport_flat = extract_flat_passport_info(text_unit.get('passport', {}))
            record.update(passport_flat)

            text_flat = extract_flat_text_content(text_unit.get('textContent', {}))

            record.update({
                'sentence_count': text_flat['sentence_count'],
                'word_count': text_flat['word_count'],
                'all_translations': text_flat['all_translations_combined'],
                'all_hieroglyphic_text': text_flat['all_hieroglyphic_combined'],
            })

            records.append(record)

        return records

    except Exception as e:
        print(f"ĞÑˆĞ¸Ğ±ĞºĞ° Ğ¾Ğ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ¸ Ñ„Ğ°Ğ¹Ğ»Ğ° {file_path}: {e}")
        return []

def main_flat_processing():
    """ĞÑĞ½Ğ¾Ğ²Ğ½Ğ°Ñ Ñ„ÑƒĞ½ĞºÑ†Ğ¸Ñ Ğ´Ğ»Ñ Ğ¿Ğ»Ğ¾ÑĞºĞ¾Ğ¹ Ğ¾Ğ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ¸"""
    base_path = "/content/drive/MyDrive/Ğ”Ñ€ĞµĞ²Ğ½Ğ¸Ğµ ÑĞ·Ñ‹ĞºĞ¸/corpus/"
    output_dir = "/content/drive/MyDrive/Ğ”Ñ€ĞµĞ²Ğ½Ğ¸Ğµ ÑĞ·Ñ‹ĞºĞ¸/corpus_export_flat/"

    os.makedirs(output_dir, exist_ok=True)

    json_files = []
    for root, dirs, files in os.walk(base_path):
        for file in files:
            if file.endswith('.json'):
                json_files.append(os.path.join(root, file))

    print(f"ĞĞ°Ğ¹Ğ´ĞµĞ½Ğ¾ JSON Ñ„Ğ°Ğ¹Ğ»Ğ¾Ğ²: {len(json_files)}")

    if not json_files:
        print(" Ğ¤Ğ°Ğ¹Ğ»Ñ‹ Ğ½Ğµ Ğ½Ğ°Ğ¹Ğ´ĞµĞ½Ñ‹")
        return

    BATCH_SIZE = 3
    all_records = []

    for batch_num in range(0, len(json_files), BATCH_SIZE):
        batch_files = json_files[batch_num:batch_num + BATCH_SIZE]
        batch_records = []

        print(f"\n=== ĞŸĞ°ĞºĞµÑ‚ {batch_num//BATCH_SIZE + 1}/{(len(json_files)-1)//BATCH_SIZE + 1} ===")

        for file_path in batch_files:
            print(f"ĞĞ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ°: {os.path.basename(file_path)}")
            records = process_single_file_flat(file_path)
            batch_records.extend(records)

        if batch_records:
            batch_df = pd.DataFrame(batch_records)
            batch_path = os.path.join(output_dir, f"batch_flat_{batch_num//BATCH_SIZE + 1:04d}.parquet")
            batch_df.to_parquet(batch_path, index=False)
            print(f" ĞŸĞ°ĞºĞµÑ‚ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½: {len(batch_records)} Ğ·Ğ°Ğ¿Ğ¸ÑĞµĞ¹, {len(batch_df.columns)} ĞºĞ¾Ğ»Ğ¾Ğ½Ğ¾Ğº")

            all_records.extend(batch_records)

        gc.collect()

    if all_records:
        print(f"\n=== ĞĞ±ÑŠĞµĞ´Ğ¸Ğ½ĞµĞ½Ğ¸Ğµ Ğ¿Ğ°ĞºĞµÑ‚Ğ¾Ğ² ===")
        final_df = pd.DataFrame(all_records)

        final_parquet = os.path.join(output_dir, "corpus_completely_flat.parquet")
        final_csv = os.path.join(output_dir, "corpus_completely_flat.csv")

        final_df.to_parquet(final_parquet, index=False)
        final_df.to_csv(final_csv, index=False, encoding='utf-8')

        print(f"Ğ¤Ğ¸Ğ½Ğ°Ğ»ÑŒĞ½Ñ‹Ğµ Ñ„Ğ°Ğ¹Ğ»Ñ‹ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ñ‹:")
        print(f"   - {final_parquet}")
        print(f"   - {final_csv}")
        print(f"Ğ˜Ñ‚Ğ¾Ğ³Ğ¾Ğ²Ğ°Ñ ÑÑ‚Ğ°Ñ‚Ğ¸ÑÑ‚Ğ¸ĞºĞ°:")
        print(f"   - Ğ—Ğ°Ğ¿Ğ¸ÑĞµĞ¹: {len(final_df)}")
        print(f"   - ĞšĞ¾Ğ»Ğ¾Ğ½Ğ¾Ğº: {len(final_df.columns)}")
        print(f"   - ĞŸÑ€Ğ¸Ğ¼ĞµÑ€ ĞºĞ¾Ğ»Ğ¾Ğ½Ğ¾Ğº: {list(final_df.columns)[:10]}...")

        columns_info = []
        for col in final_df.columns:
            non_null = final_df[col].notna().sum()
            columns_info.append(f"{col}: {non_null} Ğ½Ğµ-Ğ¿ÑƒÑÑ‚Ñ‹Ñ… Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ¸Ğ¹")

        with open(os.path.join(output_dir, "columns_description.txt"), 'w', encoding='utf-8') as f:
            f.write("ĞĞ¿Ğ¸ÑĞ°Ğ½Ğ¸Ğµ ĞºĞ¾Ğ»Ğ¾Ğ½Ğ¾Ğº Ğ¿Ğ¾Ğ»Ğ½Ğ¾ÑÑ‚ÑŒÑ Ñ€Ğ°ÑĞ¿Ğ°ĞºĞ¾Ğ²Ğ°Ğ½Ğ½Ğ¾Ğ¹ Ñ‚Ğ°Ğ±Ğ»Ğ¸Ñ†Ñ‹:\n")
            f.write("=" * 50 + "\n")
            f.write("\n".join(columns_info))

    return final_df if 'final_df' in locals() else None

def create_normalized_tables(flat_df, output_dir):
    """Ğ¡Ğ¾Ğ·Ğ´Ğ°ĞµÑ‚ Ğ½Ğ¾Ñ€Ğ¼Ğ°Ğ»Ğ¸Ğ·Ğ¾Ğ²Ğ°Ğ½Ğ½Ñ‹Ğµ Ñ‚Ğ°Ğ±Ğ»Ğ¸Ñ†Ñ‹ Ğ¸Ğ· Ğ¿Ğ»Ğ¾ÑĞºĞ¸Ñ… Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…"""

    main_columns = [col for col in flat_df.columns if not any(x in col for x in ['sentence', 'word_', 'translation'])]
    main_table = flat_df[main_columns].copy()

    text_columns = ['text_unit_id', 'all_translations', 'all_hieroglyphic_text', 'sentence_count', 'word_count']
    text_table = flat_df[text_columns].copy()

    main_table.to_parquet(os.path.join(output_dir, "normalized_main.parquet"), index=False)
    text_table.to_parquet(os.path.join(output_dir, "normalized_text_content.parquet"), index=False)

    print(f" ĞĞ¾Ñ€Ğ¼Ğ°Ğ»Ğ¸Ğ·Ğ¾Ğ²Ğ°Ğ½Ğ½Ñ‹Ğµ Ñ‚Ğ°Ğ±Ğ»Ğ¸Ñ†Ñ‹ ÑĞ¾Ğ·Ğ´Ğ°Ğ½Ñ‹")
    print(f"   - ĞÑĞ½Ğ¾Ğ²Ğ½Ğ°Ñ Ñ‚Ğ°Ğ±Ğ»Ğ¸Ñ†Ğ°: {len(main_table)} Ğ·Ğ°Ğ¿Ğ¸ÑĞµĞ¹, {len(main_table.columns)} ĞºĞ¾Ğ»Ğ¾Ğ½Ğ¾Ğº")
    print(f"   - Ğ¢ĞµĞºÑÑ‚Ğ¾Ğ²Ğ°Ñ Ñ‚Ğ°Ğ±Ğ»Ğ¸Ñ†Ğ°: {len(text_table)} Ğ·Ğ°Ğ¿Ğ¸ÑĞµĞ¹")
```

```python
def create_id_to_name_mapping(thesaurus_data):
    """Ğ¡Ğ¾Ğ·Ğ´Ğ°ĞµÑ‚ ÑĞ»Ğ¾Ğ²Ğ°Ñ€ÑŒ Ğ´Ğ»Ñ Ğ¿ĞµÑ€ĞµĞ²Ğ¾Ğ´Ğ° ID Ğ² Ñ‡Ğ¸Ñ‚Ğ°ĞµĞ¼Ñ‹Ğµ Ğ½Ğ°Ğ·Ğ²Ğ°Ğ½Ğ¸Ñ"""
    mapping = {}
    for entry in thesaurus_data:
        entry_id = entry.get('_id')
        entry_name = entry.get('name')
        if entry_id and entry_name:
            mapping[entry_id] = entry_name
    return mapping


id_to_name = create_id_to_name_mapping(thesaurus)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

PERIOD_TO_KINGDOM = {
    # Ğ”Ğ¾Ğ´Ğ¸Ğ½Ğ°ÑÑ‚Ğ¸Ñ‡ĞµÑĞºĞ¸Ğ¹ Ğ¿ĞµÑ€Ğ¸Ğ¾Ğ´ -5300 Ğ´Ğ¾ -3032
    'PrÃ¤dynastische Zeit': 'PrÃ¤dynastische Zeit',
    'PrÃ¤dynastisch': 'PrÃ¤dynastische Zeit',
    'Naqada': 'PrÃ¤dynastische Zeit',

    # Ğ Ğ°Ğ½Ğ½ĞµĞ´Ğ¸Ğ½Ğ°ÑÑ‚Ğ¸Ñ‡ĞµÑĞºĞ¸Ğ¹ Ğ¿ĞµÑ€Ğ¸Ğ¾Ğ´ -3032 Ğ´Ğ¾ -2686
    'FrÃ¼hdynastische Zeit': 'FrÃ¼hdynastische Zeit',
    '0. Dyn.': 'FrÃ¼hdynastische Zeit',
    '1. Dyn.': 'FrÃ¼hdynastische Zeit',
    '2. Dyn.': 'FrÃ¼hdynastische Zeit',

    # Ğ”Ñ€ĞµĞ²Ğ½ĞµĞµ Ñ†Ğ°Ñ€ÑÑ‚Ğ²Ğ¾ (Altes Reich) -2686 Ğ´Ğ¾ -2181
    'Altes Reich': 'Altes Reich',
    '3. Dyn.': 'Altes Reich',
    '4. Dyn.': 'Altes Reich',
    '5. Dyn.': 'Altes Reich',
    '6. Dyn.': 'Altes Reich',
    '6. Dyn. nach Pepi II.': 'Altes Reich',
    'Djoser / Netjeri-chet': 'Altes Reich',
    'Djeserteti / Sechemchet': 'Altes Reich',
    'Nebka / Sanacht': 'Altes Reich',
    'Snofru': 'Altes Reich',
    'Cheops': 'Altes Reich',
    'Djedefre': 'Altes Reich',
    'Chefren': 'Altes Reich',
    'Chephren': 'Altes Reich',
    'Mykerinos': 'Altes Reich',
    'Schepseskaf': 'Altes Reich',
    'Userkaf': 'Altes Reich',
    'Userkare': 'Altes Reich',
    'Sahure': 'Altes Reich',
    'Neferirkare': 'Altes Reich',
    'Niuserre': 'Altes Reich',
    'Menkauhor': 'Altes Reich',
    'Djedkare': 'Altes Reich',
    'Djedkare Asosi': 'Altes Reich',
    'Unas': 'Altes Reich',
    'Teti': 'Altes Reich',
    'Pepi I.': 'Altes Reich',
    'Pepi II.': 'Altes Reich',
    'Merenre': 'Altes Reich',
    'Tamphthis': 'Altes Reich',

    # ĞŸĞµÑ€Ğ²Ñ‹Ğ¹ Ğ¿ĞµÑ€ĞµÑ…Ğ¾Ğ´Ğ½Ñ‹Ğ¹ Ğ¿ĞµÑ€Ğ¸Ğ¾Ğ´ -2181 Ğ´Ğ¾ -2055
    '1. Zwischenzeit': '1. Zwischenzeit',
    'Erste Zwischenzeit': '1. Zwischenzeit',
    '7. Dyn.': '1. Zwischenzeit',
    '8. Dyn.': '1. Zwischenzeit',
    '9. Dyn.': '1. Zwischenzeit',
    '10. Dyn.': '1. Zwischenzeit',
    '9./10. Dyn. vor Beginn der 11. Dyn.': '1. Zwischenzeit',
    '9./10. Dyn. parallel zur frÃ¼hen 11. Dyn.': '1. Zwischenzeit',
    'Herakleopolitenzeit (Gesamtzeitraum)': '1. Zwischenzeit',
    '11. Dyn.': '1. Zwischenzeit',  # Ğ Ğ°Ğ½Ğ½ÑÑ 11 Ğ´Ğ¸Ğ½Ğ°ÑÑ‚Ğ¸Ñ ĞµÑ‰Ğµ Ğ¿ĞµÑ€ĞµÑ…Ğ¾Ğ´Ğ½Ñ‹Ğ¹ Ğ¿ĞµÑ€Ğ¸Ğ¾Ğ´

    # Ğ¡Ñ€ĞµĞ´Ğ½ĞµĞµ Ñ†Ğ°Ñ€ÑÑ‚Ğ²Ğ¾ (Mittleres Reich) -2055 Ğ´Ğ¾ -1650
    'Mittleres Reich': 'Mittleres Reich',
    '11. Dyn., 2. HÃ¤lfte': 'Mittleres Reich',
    '11. Dyn. (Gesamtzeitraum)': 'Mittleres Reich',
    '12. Dyn.': 'Mittleres Reich',
    '13. Dyn.': 'Mittleres Reich',
    'Mentuhotep II.': 'Mittleres Reich',
    'Mentuhotep II. Nebhepetre': 'Mittleres Reich',
    'Mentuhotep II. Nebhepetre (vor der Reichseinigung)': 'Mittleres Reich',
    'Mentuhotep II. Nebhepetre (nach der Reichseinigung)': 'Mittleres Reich',
    'Intef II. Wahanch': 'Mittleres Reich',
    'Intef III. Nachtnebtepnefer': 'Mittleres Reich',
    'Amenemhet I.': 'Mittleres Reich',
    'Sesostris I.': 'Mittleres Reich',
    'Amenemhet II.': 'Mittleres Reich',
    'Sesostris II.': 'Mittleres Reich',
    'Sesostris III.': 'Mittleres Reich',
    'Amenemhet III.': 'Mittleres Reich',
    'Amenemhet IV.': 'Mittleres Reich',
    'Sebekhotep IV.': 'Mittleres Reich',

    # Ğ’Ñ‚Ğ¾Ñ€Ğ¾Ğ¹ Ğ¿ĞµÑ€ĞµÑ…Ğ¾Ğ´Ğ½Ñ‹Ğ¹ Ğ¿ĞµÑ€Ğ¸Ğ¾Ğ´ -1650 Ğ´Ğ¾ -1550
    '2. Zwischenzeit': '2. Zwischenzeit',
    'Zweite Zwischenzeit': '2. Zwischenzeit',
    '14. Dyn.': '2. Zwischenzeit',
    '15. Dyn.': '2. Zwischenzeit',
    '16. Dyn.': '2. Zwischenzeit',
    '17. Dyn.': '2. Zwischenzeit',
    '13. / 14. Dynastie': '2. Zwischenzeit',
    '15. / 16. / 17. Dynastie': '2. Zwischenzeit',
    'Hyksos': '2. Zwischenzeit',

    # ĞĞ¾Ğ²Ğ¾Ğµ Ñ†Ğ°Ñ€ÑÑ‚Ğ²Ğ¾ (Neues Reich) -1550 Ğ´Ğ¾ -1077
    'Neues Reich': 'Neues Reich',
    '18. Dyn.': 'Neues Reich',
    '19. Dyn.': 'Neues Reich',
    '20. Dyn.': 'Neues Reich',
    'Ahmose': 'Neues Reich',
    'Amenophis I.': 'Neues Reich',
    'Thutmosis I.': 'Neues Reich',
    'Thutmosis II.': 'Neues Reich',
    'Hatschepsut': 'Neues Reich',
    'Thutmosis III.': 'Neues Reich',
    'Thutmosis III. (Gesamtzeitraum)': 'Neues Reich',
    'Amenophis II.': 'Neues Reich',
    'Thutmosis IV.': 'Neues Reich',
    'Amenophis III.': 'Neues Reich',
    'Amenophis IV. / Echnaton': 'Neues Reich',
    'Echnaton': 'Neues Reich',
    'Tutanchamun': 'Neues Reich',
    'Eje': 'Neues Reich',
    'Haremhab': 'Neues Reich',
    'Ramses I.': 'Neues Reich',
    'Sethos I.': 'Neues Reich',
    'Ramses II.': 'Neues Reich',
    'Merenptah': 'Neues Reich',
    'Sethos II.': 'Neues Reich',
    'Siptah': 'Neues Reich',
    'Siptah u. Tausret': 'Neues Reich',
    'Ramses III.': 'Neues Reich',
    'Ramses IV.': 'Neues Reich',
    'Ramses V.': 'Neues Reich',
    'Ramses VI.': 'Neues Reich',
    'Ramses VII.': 'Neues Reich',
    'Ramses IX.': 'Neues Reich',
    'Ramses XI.': 'Neues Reich',
    'Ramessiden': 'Neues Reich',
    'Amenophis I.': 'Neues Reich',
    'Thutmosis I.': 'Neues Reich',
    'Thutmosis II.': 'Neues Reich',
    'Hatschepsut': 'Neues Reich',
    'Thutmosis III.': 'Neues Reich',
    'Amenophis II.': 'Neues Reich',
    'Thutmosis IV.': 'Neues Reich',
    'Amenophis III.': 'Neues Reich',
    'Echnaton': 'Neues Reich',
    'Tutanchamun': 'Neues Reich',
    'Eje': 'Neues Reich',
    'Haremhab': 'Neues Reich',
    'Ramses I.': 'Neues Reich',
    'Sethos I.': 'Neues Reich',
    'Ramses II.': 'Neues Reich',
    'Merenptah': 'Neues Reich',
    'Sethos II.': 'Neues Reich',
    'Siptah': 'Neues Reich',
    'Ramses III.': 'Neues Reich',
    'Ramses IV.': 'Neues Reich',
    'Ramses VI.': 'Neues Reich',
    'Ramses IX.': 'Neues Reich',
    'Ramses XI.': 'Neues Reich',
    'Ramessiden': 'Neues Reich',

    # Ğ¢Ñ€ĞµÑ‚Ğ¸Ğ¹ Ğ¿ĞµÑ€ĞµÑ…Ğ¾Ğ´Ğ½Ñ‹Ğ¹ Ğ¿ĞµÑ€Ğ¸Ğ¾Ğ´ -1077 Ğ´Ğ¾ -664
    '3. Zwischenzeit': '3. Zwischenzeit',
    'Dritte Zwischenzeit': '3. Zwischenzeit',
    '21. Dyn.': '3. Zwischenzeit',
    '22. Dyn.': '3. Zwischenzeit',
    '22. / 23. Dyn.': '3. Zwischenzeit',
    '23. Dyn.': '3. Zwischenzeit',
    '24. Dyn.': '3. Zwischenzeit',
    '25. Dyn.': '3. Zwischenzeit',
    'Siamun': '3. Zwischenzeit',
    'Pije': '3. Zwischenzeit',
    'Taharqa': '3. Zwischenzeit',
    'Tanutamon': '3. Zwischenzeit',

    # ĞŸĞ¾Ğ·Ğ´Ğ½Ğ¸Ğ¹ Ğ¿ĞµÑ€Ğ¸Ğ¾Ğ´ (SpÃ¤tzeit) -664 Ğ´Ğ¾ -332
    'SpÃ¤tzeit': 'SpÃ¤tzeit',
    '26. Dyn.': 'SpÃ¤tzeit',
    '27. Dyn.': 'SpÃ¤tzeit',
    '28. Dyn.': 'SpÃ¤tzeit',
    '29. Dyn.': 'SpÃ¤tzeit',
    '30. Dyn.': 'SpÃ¤tzeit',
    'Psammetich I.': 'SpÃ¤tzeit',
    'Amasis': 'SpÃ¤tzeit',
    'Apries': 'SpÃ¤tzeit',

    # Ğ“Ñ€ĞµĞºĞ¾-Ñ€Ğ¸Ğ¼ÑĞºĞ¸Ğ¹ Ğ¿ĞµÑ€Ğ¸Ğ¾Ğ´ -332 Ğ´Ğ¾ 395
    'PtolemÃ¤er': 'Griechisch-RÃ¶mische Zeit',
    'PtolemÃ¤erzeit': 'Griechisch-RÃ¶mische Zeit',
    'Makedonen, PtolemÃ¤er': 'Griechisch-RÃ¶mische Zeit',
    'Griechisch-rÃ¶mische Zeit': 'Griechisch-RÃ¶mische Zeit',
    'Griechisch-RÃ¶mische Zeit': 'Griechisch-RÃ¶mische Zeit',
    'RÃ¶mische Kaiserzeit': 'Griechisch-RÃ¶mische Zeit',
    'rÃ¶mische Zeit': 'Griechisch-RÃ¶mische Zeit',
    'rÃ¶misch': 'Griechisch-RÃ¶mische Zeit',
}

KINGDOM_DATES = {
    'PrÃ¤dynastische Zeit': (-5300, -3032),
    'FrÃ¼hdynastische Zeit': (-3032, -2686),
    'Altes Reich': (-2686, -2181),
    '1. Zwischenzeit': (-2181, -2055),
    'Mittleres Reich': (-2055, -1650),
    '2. Zwischenzeit': (-1650, -1550),
    'Neues Reich': (-1550, -1077),
    '3. Zwischenzeit': (-1077, -664),
    'SpÃ¤tzeit': (-664, -332),
    'Griechisch-RÃ¶mische Zeit': (-332, 395)
}


def parse_german_century_to_kingdom(text):
    """ĞŸĞ°Ñ€ÑĞ¸Ñ‚ Ğ²ĞµĞºĞ° Ğ¸ Ğ¾Ğ¿Ñ€ĞµĞ´ĞµĞ»ÑĞµÑ‚ Ñ†Ğ°Ñ€ÑÑ‚Ğ²Ğ¾"""
    if pd.isna(text):
        return None

    text = str(text).lower()

    if 'jhdt. v.chr.' in text or 'jh. v.chr.' in text:
        import re
        match = re.search(r'(\d+)\.', text)
        if match:
            century = int(match.group(1))

            if century >= 20:  # 20+ Ğ²ĞµĞº Ğ´Ğ¾ Ğ½.Ñ.
                return 'Altes Reich'
            elif century >= 16:  # 16-19 Ğ²ĞµĞº Ğ´Ğ¾ Ğ½.Ñ.
                return 'Mittleres Reich'
            elif century >= 11:  # 11-15 Ğ²ĞµĞº Ğ´Ğ¾ Ğ½.Ñ.
                return 'Neues Reich'
            elif century >= 7:  # 7-10 Ğ²ĞµĞº Ğ´Ğ¾ Ğ½.Ñ.
                return '3. Zwischenzeit'
            elif century >= 4:  # 4-6 Ğ²ĞµĞº Ğ´Ğ¾ Ğ½.Ñ.
                return 'SpÃ¤tzeit'
            else:  # 1-3 Ğ²ĞµĞº Ğ´Ğ¾ Ğ½.Ñ.
                return 'Griechisch-RÃ¶mische Zeit'

    return None


def assign_kingdom(period_text):
    """ĞĞ¿Ñ€ĞµĞ´ĞµĞ»ÑĞµÑ‚ Ñ†Ğ°Ñ€ÑÑ‚Ğ²Ğ¾ Ğ´Ğ»Ñ Ğ¿ĞµÑ€Ğ¸Ğ¾Ğ´Ğ°"""
    if pd.isna(period_text):
        return None

    period_text = str(period_text).strip()

    # ĞŸÑ€ÑĞ¼Ğ¾Ğµ ÑĞ¾Ğ²Ğ¿Ğ°Ğ´ĞµĞ½Ğ¸Ğµ Ğ² ÑĞ»Ğ¾Ğ²Ğ°Ñ€Ğµ
    if period_text in PERIOD_TO_KINGDOM:
        return PERIOD_TO_KINGDOM[period_text]

    # ĞŸÑ€Ğ¾Ğ²ĞµÑ€ÑĞµĞ¼ Ñ‡Ğ°ÑÑ‚Ğ¸Ñ‡Ğ½Ñ‹Ğµ ÑĞ¾Ğ²Ğ¿Ğ°Ğ´ĞµĞ½Ğ¸Ñ
    period_lower = period_text.lower()
    for key, kingdom in PERIOD_TO_KINGDOM.items():
        if key.lower() in period_lower:
            return kingdom

    # ĞŸÑ€Ğ¾Ğ±ÑƒĞµĞ¼ Ğ¿Ğ°Ñ€ÑĞ¸Ñ‚ÑŒ Ğ²ĞµĞºĞ°
    kingdom = parse_german_century_to_kingdom(period_text)
    if kingdom:
        return kingdom

    return 'ĞĞµĞ¸Ğ·Ğ²ĞµÑÑ‚Ğ½Ğ¾'


def assign_kingdom_from_dates(date_start, date_end):
    """ĞĞ¿Ñ€ĞµĞ´ĞµĞ»ÑĞµÑ‚ Ñ†Ğ°Ñ€ÑÑ‚Ğ²Ğ¾ Ğ¿Ğ¾ Ñ‡Ğ¸ÑĞ»Ğ¾Ğ²Ñ‹Ğ¼ Ğ´Ğ°Ñ‚Ğ°Ğ¼ (Ğ´Ğ»Ñ HF Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚Ğ°)"""
    if pd.isna(date_start) or pd.isna(date_end):
        return None


    mid_date = (date_start + date_end) / 2


    for kingdom, (start, end) in KINGDOM_DATES.items():
        if start <= mid_date <= end:
            return kingdom


    if mid_date < -5300:
        return 'Ğ”Ğ¾ PrÃ¤dynastische Zeit'


    if mid_date > 395:
        return 'ĞŸĞ¾ÑĞ»Ğµ Griechisch-RÃ¶mische Zeit'

    return 'ĞĞµĞ¸Ğ·Ğ²ĞµÑÑ‚Ğ½Ğ¾'


def analyze_kingdoms(df_thesaurus, df_hf):
    """ĞĞ½Ğ°Ğ»Ğ¸Ğ·Ğ¸Ñ€ÑƒĞµÑ‚ Ñ€Ğ°ÑĞ¿Ñ€ĞµĞ´ĞµĞ»ĞµĞ½Ğ¸Ğµ Ğ¿Ğ¾ Ñ†Ğ°Ñ€ÑÑ‚Ğ²Ğ°Ğ¼"""


    df_thes = df_thesaurus.copy()
    df_thes['kingdom'] = df_thes['period'].apply(assign_kingdom)


    df_hf = df_hf.copy()
    df_hf['kingdom'] = df_hf.apply(
        lambda row: assign_kingdom_from_dates(row['dateNotBefore'], row['dateNotAfter']),
        axis=1
    )


    return df_thes, df_hf


def visualize_kingdoms(df_thes, df_hf):
    """Ğ’Ğ¸Ğ·ÑƒĞ°Ğ»Ğ¸Ğ·Ğ¸Ñ€ÑƒĞµÑ‚ Ñ€Ğ°ÑĞ¿Ñ€ĞµĞ´ĞµĞ»ĞµĞ½Ğ¸Ğµ Ğ¿Ğ¾ Ñ†Ğ°Ñ€ÑÑ‚Ğ²Ğ°Ğ¼"""

    kingdom_order = [
        'Ğ”Ğ¾ PrÃ¤dynastische Zeit',
        'PrÃ¤dynastische Zeit',
        'FrÃ¼hdynastische Zeit',
        'Altes Reich',
        '1. Zwischenzeit',
        'Mittleres Reich',
        '2. Zwischenzeit',
        'Neues Reich',
        '3. Zwischenzeit',
        'SpÃ¤tzeit',
        'Griechisch-RÃ¶mische Zeit',
        'ĞŸĞ¾ÑĞ»Ğµ Griechisch-RÃ¶mische Zeit',
        'ĞĞµĞ¸Ğ·Ğ²ĞµÑÑ‚Ğ½Ğ¾'
    ]



    thes_counts = df_thes['kingdom'].value_counts()
    thes_counts = thes_counts.reindex([k for k in kingdom_order if k in thes_counts.index], fill_value=0)


    hf_counts = df_hf['kingdom'].value_counts()
    hf_counts = hf_counts.reindex([k for k in kingdom_order if k in hf_counts.index], fill_value=0)

    fig, ax = plt.subplots(figsize=(14, 8))

    all_kingdoms = sorted(set(thes_counts.index) | set(hf_counts.index))
    x = np.arange(len(all_kingdoms))
    width = 0.35

    thes_values = [thes_counts.get(k, 0) for k in all_kingdoms]
    hf_values = [hf_counts.get(k, 0) for k in all_kingdoms]

    ax.barh(x - width/2, thes_values, width, label='Ğ¢ĞµĞ·Ğ°ÑƒÑ€ÑƒÑ', color='steelblue')
    ax.barh(x + width/2, hf_values, width, label='HF Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚', color='coral')

    ax.set_yticks(x)
    ax.set_yticklabels(all_kingdoms)
    ax.set_xlabel('ĞšĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ Ñ‚ĞµĞºÑÑ‚Ğ¾Ğ²')
    ax.set_title('Ğ¡Ñ€Ğ°Ğ²Ğ½ĞµĞ½Ğ¸Ğµ Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚Ğ¾Ğ² Ğ¿Ğ¾ Ñ†Ğ°Ñ€ÑÑ‚Ğ²Ğ°Ğ¼', fontsize=14, fontweight='bold')
    ax.legend()
    ax.grid(axis='x', alpha=0.3)

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import re
from pathlib import Path

def clean_transliteration(text, source='thesaurus'):
    """
    ĞÑ‡Ğ¸Ñ‰Ğ°ĞµÑ‚ Ñ‚Ñ€Ğ°Ğ½ÑĞ»Ğ¸Ñ‚ĞµÑ€Ğ°Ñ†Ğ¸Ñ Ğ¾Ñ‚ Ñ‚ĞµÑ…Ğ½Ğ¸Ñ‡ĞµÑĞºĞ¸Ñ… ÑĞ¸Ğ¼Ğ²Ğ¾Ğ»Ğ¾Ğ².

    Parameters:
    -----------
    text : str
        Ğ¢Ñ€Ğ°Ğ½ÑĞ»Ğ¸Ñ‚ĞµÑ€Ğ°Ñ†Ğ¸Ñ
    source : str
        'thesaurus' Ğ¸Ğ»Ğ¸ 'hf' - Ñ€Ğ°Ğ·Ğ½Ñ‹Ğµ Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ‚Ñ‹
    """
    if pd.isna(text) or text == '':
        return ''

    text = str(text)

    # Ğ¿Ğ¾ĞºĞ° ÑƒĞ±Ğ¸Ñ€Ğ°Ñ, Ğ½Ğ°Ğ´Ğ¾ Ğ¿Ğ¾Ğ¿Ñ€Ğ°Ğ²Ğ¸Ñ‚ÑŒ!!!
    text = re.sub(r'â¸¢|â¸£|\[|\]|\{|\}|ã€ˆ|ã€‰', '', text)

    if source == 'thesaurus':
        text = text.replace(' | ', ' ')
        text = text.replace('|', ' ')

    text = re.sub(r'\s+', ' ', text).strip()

    return text


def normalize_tokens(text):
    """
    ĞĞ¾Ñ€Ğ¼Ğ°Ğ»Ğ¸Ğ·ÑƒĞµÑ‚ Ñ‚Ğ¾ĞºĞµĞ½Ñ‹ Ğ´Ğ»Ñ ĞµĞ´Ğ¸Ğ½Ğ¾Ğ¾Ğ±Ñ€Ğ°Ğ·Ğ¸Ñ.
    - Ğ£Ğ±Ğ¸Ñ€Ğ°ĞµÑ‚ =, (, ) Ğ¸ Ğ´Ñ€ÑƒĞ³Ğ¸Ğµ Ñ‚ĞµÑ…Ğ½Ğ¸Ñ‡ĞµÑĞºĞ¸Ğµ ÑĞ¸Ğ¼Ğ²Ğ¾Ğ»Ñ‹
    """
    text = text.replace('=', '')
    text = re.sub(r'[()]', '', text)

    return text


def prepare_altes_reich_corpus(df_thes, df_hf, output_path,
                                normalize=True,
                                min_words=2):
    """
    ĞŸĞ¾Ğ´Ğ³Ğ¾Ñ‚Ğ°Ğ²Ğ»Ğ¸Ğ²Ğ°ĞµÑ‚ Ğ¾Ğ±ÑŠĞµĞ´Ğ¸Ğ½Ñ‘Ğ½Ğ½Ñ‹Ğ¹ ĞºĞ¾Ñ€Ğ¿ÑƒÑ Altes Reich Ğ´Ğ»Ñ Word2Vec.

    Parameters:
    -----------
    df_thes : DataFrame
        DataFrame Ğ¸Ğ· Ñ‚ĞµĞ·Ğ°ÑƒÑ€ÑƒÑĞ° Ñ ĞºĞ¾Ğ»Ğ¾Ğ½ĞºĞ¾Ğ¹ 'all_hieroglyphic_text'
    df_hf : DataFrame
        DataFrame Ğ¸Ğ· HF Ñ ĞºĞ¾Ğ»Ğ¾Ğ½ĞºĞ¾Ğ¹ 'transliteration'
    output_path : str
        ĞŸÑƒÑ‚ÑŒ Ğ´Ğ»Ñ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ¸Ñ
    normalize : bool
        ĞĞ¾Ñ€Ğ¼Ğ°Ğ»Ğ¸Ğ·Ğ¾Ğ²Ğ°Ñ‚ÑŒ Ñ‚Ğ¾ĞºĞµĞ½Ñ‹ (ÑƒĞ±Ñ€Ğ°Ñ‚ÑŒ =, ÑĞºĞ¾Ğ±ĞºĞ¸ Ğ¸ Ñ‚.Ğ´.)
    min_words : int
        ĞœĞ¸Ğ½Ğ¸Ğ¼Ğ°Ğ»ÑŒĞ½Ğ¾Ğµ ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ÑĞ»Ğ¾Ğ² Ğ² ÑÑ‚Ñ€Ğ¾ĞºĞµ
    """

    all_texts = []

    print("ĞĞ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ° Ñ‚ĞµĞ·Ğ°ÑƒÑ€ÑƒÑĞ°...")
    for text in df_thes['all_hieroglyphic_text']:
        cleaned = clean_transliteration(text, source='thesaurus')
        if normalize:
            cleaned = normalize_tokens(cleaned)

        if len(cleaned.split()) >= min_words:
            all_texts.append(cleaned)

    print(f"Ğ¢ĞµĞ·Ğ°ÑƒÑ€ÑƒÑ: {len(all_texts)} ÑÑ‚Ñ€Ğ¾Ğº")

    print("ĞĞ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ° HF Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚Ğ°...")
    hf_count_before = len(all_texts)
    for text in df_hf['transliteration']:
        cleaned = clean_transliteration(text, source='hf')
        if normalize:
            cleaned = normalize_tokens(cleaned)

        if len(cleaned.split()) >= min_words:
            all_texts.append(cleaned)

    hf_added = len(all_texts) - hf_count_before
    print(f"HF: {hf_added} ÑÑ‚Ñ€Ğ¾Ğº")

    print("\n=== Ğ¡Ğ¢ĞĞ¢Ğ˜Ğ¡Ğ¢Ğ˜ĞšĞ ĞšĞĞ ĞŸĞ£Ğ¡Ğ ===")
    print(f"Ğ’ÑĞµĞ³Ğ¾ ÑÑ‚Ñ€Ğ¾Ğº: {len(all_texts)}")

    all_words = []
    for text in all_texts:
        words = text.split()
        all_words.extend(words)

    print(f"Ğ’ÑĞµĞ³Ğ¾ Ñ‚Ğ¾ĞºĞµĞ½Ğ¾Ğ²: {len(all_words)}")
    print(f"Ğ£Ğ½Ğ¸ĞºĞ°Ğ»ÑŒĞ½Ñ‹Ñ… Ñ‚Ğ¾ĞºĞµĞ½Ğ¾Ğ²: {len(set(all_words))}")

    lengths = [len(text.split()) for text in all_texts]
    print(f"Ğ¡Ñ€ĞµĞ´Ğ½ÑÑ Ğ´Ğ»Ğ¸Ğ½Ğ° ÑÑ‚Ñ€Ğ¾ĞºĞ¸: {sum(lengths)/len(lengths):.1f} ÑĞ»Ğ¾Ğ²")
    print(f"Min: {min(lengths)}, Max: {max(lengths)}")

    from collections import Counter
    word_freq = Counter(all_words)
    print("\nĞ¢Ğ¾Ğ¿-20 ÑĞ°Ğ¼Ñ‹Ñ… Ñ‡Ğ°ÑÑ‚Ñ‹Ñ… ÑĞ»Ğ¾Ğ²:")
    for word, count in word_freq.most_common(20):
        print(f"  {word}: {count}")

    with open(output_path, 'w', encoding='utf-8') as f:
        for text in all_texts:
            f.write(text + '\n')

    print(f" Ğ¡Ğ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ¾ Ğ²: {output_path}")

    return all_texts


def show_examples(texts, n=10):
    """ĞŸĞ¾ĞºĞ°Ğ·Ñ‹Ğ²Ğ°ĞµÑ‚ Ğ¿Ñ€Ğ¸Ğ¼ĞµÑ€Ñ‹ Ğ¸Ğ· ĞºĞ¾Ñ€Ğ¿ÑƒÑĞ°"""
    print(f"\n=== ĞŸĞ Ğ˜ĞœĞ•Ğ Ğ« Ğ˜Ğ— ĞšĞĞ ĞŸĞ£Ğ¡Ğ (Ğ¿ĞµÑ€Ğ²Ñ‹Ğµ {n}) ===")
    for i, text in enumerate(texts[:n], 1):
        preview = text[:100] + "..." if len(text) > 100 else text
        print(f"{i:2d}. {preview}")
```

```python
def extract_lemmatized_sentences(json_file, lemma_map, text_ids_filter):

    records = []

    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    for text_unit in data:

        text_id = text_unit.get('_id', '')

        if text_ids_filter and text_id not in text_ids_filter:
            continue

        text_content = text_unit.get('textContent', {})

        for sentence in text_content.get('textItems', []):
            sentence_id = sentence.get('_id', '')

            lemmas = []
            lemma_ids = []
            surface_forms = []
            flexion_codes = []

            for item in sentence.get('sentenceItems', []):

                if 'BTSWord' not in item.get('eClass', ''):
                    continue

                lemma_id = item.get('lKey', '')
                surface = item.get('wChar', '')
                flexion = item.get('flexCode', '')

                lemma = lemma_map.get(lemma_id, surface)

                if lemma:
                    lemmas.append(lemma)
                    lemma_ids.append(lemma_id)
                    surface_forms.append(surface)
                    flexion_codes.append(flexion)

            if lemmas:
                records.append({
                    'text_id': text_id,
                    'sentence_id': sentence_id,
                    'lemmas': lemmas,
                    'lemma_ids': lemma_ids,
                    'surface_forms': surface_forms,
                    'flexion_codes': flexion_codes
                })

    return records


def process_corpus_to_lemmas(json_files, lemma_map, output_dir):
    """
    ĞĞ±Ñ€Ğ°Ğ±Ğ°Ñ‚Ñ‹Ğ²Ğ°ĞµÑ‚ Ğ²ĞµÑÑŒ ĞºĞ¾Ñ€Ğ¿ÑƒÑ Ğ¸ ÑĞ¾Ğ·Ğ´Ğ°Ñ‘Ñ‚ Ğ»ĞµĞ¼Ğ¼Ğ°Ñ‚Ğ¸Ğ·Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ½Ñ‹Ğµ Ñ„Ğ°Ğ¹Ğ»Ñ‹.
    """
    all_records = []

    for json_file in json_files:
        print(f"ĞĞ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ°: {os.path.basename(json_file)}")
        records = extract_lemmatized_sentences(json_file, lemma_map, altes_reich_ids)
        all_records.extend(records)
        print(f"  ĞĞ°Ğ¹Ğ´ĞµĞ½Ğ¾ Ğ¿Ñ€ĞµĞ´Ğ»Ğ¾Ğ¶ĞµĞ½Ğ¸Ğ¹: {len(records)}")

    print(f"\n=== Ğ¡Ğ¢ĞĞ¢Ğ˜Ğ¡Ğ¢Ğ˜ĞšĞ ===")
    print(f"Ğ’ÑĞµĞ³Ğ¾ Ğ¿Ñ€ĞµĞ´Ğ»Ğ¾Ğ¶ĞµĞ½Ğ¸Ğ¹: {len(all_records)}")


    df = pd.DataFrame(all_records)


    df['lemmas_text'] = df['lemmas'].apply(lambda x: ' '.join(x))
    df['surface_text'] = df['surface_forms'].apply(lambda x: ' '.join(x))


    all_lemmas = [lemma for record in all_records for lemma in record['lemmas']]
    print(f"Ğ’ÑĞµĞ³Ğ¾ Ğ»ĞµĞ¼Ğ¼ (Ñ‚Ğ¾ĞºĞµĞ½Ğ¾Ğ²): {len(all_lemmas)}")
    print(f"Ğ£Ğ½Ğ¸ĞºĞ°Ğ»ÑŒĞ½Ñ‹Ñ… Ğ»ĞµĞ¼Ğ¼: {len(set(all_lemmas))}")


    lemma_freq = Counter(all_lemmas)
    print("\nĞ¢Ğ¾Ğ¿-20 ÑĞ°Ğ¼Ñ‹Ñ… Ñ‡Ğ°ÑÑ‚Ñ‹Ñ… Ğ»ĞµĞ¼Ğ¼:")
    for lemma, count in lemma_freq.most_common(20):
        print(f"  {lemma}: {count}")


    lemmas_file = output_dir / "corpus_lemmatized.txt"
    with open(lemmas_file, 'w', encoding='utf-8') as f:
        for record in all_records:
            f.write(' '.join(record['lemmas']) + '\n')
    print(f"Ğ›ĞµĞ¼Ğ¼Ñ‹ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ñ‹: {lemmas_file}")

    # 2. CSV Ñ Ğ¿Ğ¾Ğ»Ğ½Ğ¾Ğ¹ Ğ¸Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ğ¸ĞµĞ¹
    csv_file = output_dir / "corpus_lemmatized_full.csv"
    df.to_csv(csv_file, index=False, encoding='utf-8')
    print(f"ĞŸĞ¾Ğ»Ğ½Ñ‹Ğ¹ CSV: {csv_file}")

    # 3. Parquet (ÑÑ„Ñ„ĞµĞºÑ‚Ğ¸Ğ²Ğ½ĞµĞµ Ğ´Ğ»Ñ Ğ±Ğ¾Ğ»ÑŒÑˆĞ¸Ñ… Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…)
    parquet_file = output_dir / "corpus_lemmatized_full.parquet"
    df.to_parquet(parquet_file, index=False)
    print(f"Parquet: {parquet_file}")

    return df


def compare_surface_vs_lemma(df, n_examples=10):
    """
    Ğ¡Ñ€Ğ°Ğ²Ğ½Ğ¸Ğ²Ğ°ĞµÑ‚ Ğ¿Ğ¾Ğ²ĞµÑ€Ñ…Ğ½Ğ¾ÑÑ‚Ğ½Ñ‹Ğµ Ñ„Ğ¾Ñ€Ğ¼Ñ‹ Ğ¸ Ğ»ĞµĞ¼Ğ¼Ñ‹.
    """
    print("\n=== Ğ¡Ğ ĞĞ’ĞĞ•ĞĞ˜Ğ•: ĞŸĞĞ’Ğ•Ğ Ğ¥ĞĞĞ¡Ğ¢ĞĞĞ¯ Ğ¤ĞĞ ĞœĞ vs Ğ›Ğ•ĞœĞœĞ ===")
    print("(Ğ¿ĞµÑ€Ğ²Ñ‹Ğµ 10 Ğ¿Ñ€ĞµĞ´Ğ»Ğ¾Ğ¶ĞµĞ½Ğ¸Ğ¹)\n")

    for i, row in df.head(n_examples).iterrows():
        print(f"ĞŸÑ€ĞµĞ´Ğ»Ğ¾Ğ¶ĞµĞ½Ğ¸Ğµ {i+1}:")
        print(f"  ĞŸĞ¾Ğ²ĞµÑ€Ñ…Ğ½Ğ¾ÑÑ‚Ğ½Ğ°Ñ Ñ„Ğ¾Ñ€Ğ¼Ğ°: {row['surface_text']}")
        print(f"  Ğ›ĞµĞ¼Ğ¼Ğ°Ñ‚Ğ¸Ğ·Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ½Ğ°Ñ:  {row['lemmas_text']}")
        print()


def create_filtered_lemma_corpus(df, output_path,
                                 min_sentence_length=3,
                                 remove_names=True):
    """
    Ğ¡Ğ¾Ğ·Ğ´Ğ°Ñ‘Ñ‚ Ğ¾Ñ‚Ñ„Ğ¸Ğ»ÑŒÑ‚Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ½Ñ‹Ğ¹ ĞºĞ¾Ñ€Ğ¿ÑƒÑ Ğ»ĞµĞ¼Ğ¼.

    Parameters:
    -----------
    min_sentence_length : int
        ĞœĞ¸Ğ½Ğ¸Ğ¼Ğ°Ğ»ÑŒĞ½Ğ°Ñ Ğ´Ğ»Ğ¸Ğ½Ğ° Ğ¿Ñ€ĞµĞ´Ğ»Ğ¾Ğ¶ĞµĞ½Ğ¸Ñ
    remove_names : bool
        Ğ£Ğ´Ğ°Ğ»ÑÑ‚ÑŒ Ğ»Ğ¸ Ğ¸Ğ¼ĞµĞ½Ğ° ÑĞ¾Ğ±ÑÑ‚Ğ²ĞµĞ½Ğ½Ñ‹Ğµ (Ğ½Ğ°Ñ‡Ğ¸Ğ½Ğ°ÑÑ‚ÑÑ Ñ Ğ·Ğ°Ğ³Ğ»Ğ°Ğ²Ğ½Ğ¾Ğ¹ Ğ±ÑƒĞºĞ²Ñ‹)
    """
    filtered_sentences = []

    for _, row in df.iterrows():
        lemmas = row['lemmas']

        # Ğ¤Ğ¸Ğ»ÑŒÑ‚Ñ€ 1: Ğ´Ğ»Ğ¸Ğ½Ğ°
        if len(lemmas) < min_sentence_length:
            continue

        # Ğ¤Ğ¸Ğ»ÑŒÑ‚Ñ€ 2: Ğ¸Ğ¼ĞµĞ½Ğ° ÑĞ¾Ğ±ÑÑ‚Ğ²ĞµĞ½Ğ½Ñ‹Ğµ
        if remove_names:
            lemmas = [l for l in lemmas if not l[0].isupper()]

        if len(lemmas) >= min_sentence_length:
            filtered_sentences.append(' '.join(lemmas))

    print(f"\n=== Ğ¤Ğ˜Ğ›Ğ¬Ğ¢Ğ ĞĞ¦Ğ˜Ğ¯ ===")
    print(f"Ğ”Ğ¾ Ñ„Ğ¸Ğ»ÑŒÑ‚Ñ€Ğ°Ñ†Ğ¸Ğ¸: {len(df)} Ğ¿Ñ€ĞµĞ´Ğ»Ğ¾Ğ¶ĞµĞ½Ğ¸Ğ¹")
    print(f"ĞŸĞ¾ÑĞ»Ğµ Ñ„Ğ¸Ğ»ÑŒÑ‚Ñ€Ğ°Ñ†Ğ¸Ğ¸: {len(filtered_sentences)} Ğ¿Ñ€ĞµĞ´Ğ»Ğ¾Ğ¶ĞµĞ½Ğ¸Ğ¹")

    # Ğ¡Ğ¾Ñ…Ñ€Ğ°Ğ½ÑĞµĞ¼
    with open(output_path, 'w', encoding='utf-8') as f:
        for sent in filtered_sentences:
            f.write(sent + '\n')

    print(f" Ğ¡Ğ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ¾: {output_path}")

    return filtered_sentences
```

```python
def parse_hf_lemmatization(lemma_string):

    if pd.isna(lemma_string) or lemma_string == '':
        return ''

    lemmas = []
    for token in str(lemma_string).split():
        if '|' in token:
            parts = token.split('|')
            if len(parts) == 2:
                lemma = parts[1]
                lemmas.append(lemma)

    return ' '.join(lemmas)


print("ĞĞ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ° HF Ğ»ĞµĞ¼Ğ¼...")
hf_lemmas = []
for lemma_str in old_kingdom_hf['lemmatization']:
    parsed = parse_hf_lemmatization(lemma_str)
    if len(parsed.split()) >= 2:
        hf_lemmas.append(parsed)

print(f"HF Ğ»ĞµĞ¼Ğ¼Ñ‹: {len(hf_lemmas)} ÑÑ‚Ñ€Ğ¾Ğº")

```

```python
import unicodedata

def analyze_unicode_chars(text_list, source_name):
    """ĞĞ½Ğ°Ğ»Ğ¸Ğ·Ğ¸Ñ€ÑƒĞµÑ‚ Ğ²ÑĞµ ÑƒĞ½Ğ¸ĞºĞ°Ğ»ÑŒĞ½Ñ‹Ğµ Unicode ÑĞ¸Ğ¼Ğ²Ğ¾Ğ»Ñ‹ Ğ² ĞºĞ¾Ñ€Ğ¿ÑƒÑĞµ"""
    all_chars = set()
    for text in text_list:
        if pd.notna(text) and text:
            all_chars.update(str(text))

    special_chars = []
    for char in sorted(all_chars):
        code = ord(char)
        if code > 127:
            try:
                name = unicodedata.name(char)
                special_chars.append((char, code, name))
            except:
                special_chars.append((char, code, 'UNKNOWN'))

    print(f"\n=== {source_name} ===")
    print(f"Ğ’ÑĞµĞ³Ğ¾ ÑƒĞ½Ğ¸ĞºĞ°Ğ»ÑŒĞ½Ñ‹Ñ… ÑĞ¸Ğ¼Ğ²Ğ¾Ğ»Ğ¾Ğ²: {len(all_chars)}")
    print(f"Ğ¡Ğ¿ĞµÑ†Ğ¸Ğ°Ğ»ÑŒĞ½Ñ‹Ñ… ÑĞ¸Ğ¼Ğ²Ğ¾Ğ»Ğ¾Ğ² (>ASCII): {len(special_chars)}")
    print("\nĞ¡Ğ¿ĞµÑ†Ğ¸Ğ°Ğ»ÑŒĞ½Ñ‹Ğµ ÑĞ¸Ğ¼Ğ²Ğ¾Ğ»Ñ‹:")
    for char, code, name in special_chars:  # ĞŸĞ¾ĞºĞ°Ğ·Ñ‹Ğ²Ğ°ĞµĞ¼ Ğ’Ğ¡Ğ•
        print(f"  '{char}' â†’ U+{code:04X} ({code}) - {name}")

    return special_chars
```

```python
def normalize_egyptian_transliteration(text):
    """
    ĞŸÑ€Ğ¸Ğ²Ğ¾Ğ´Ğ¸Ñ‚ ĞµĞ³Ğ¸Ğ¿Ñ‚Ğ¾Ğ»Ğ¾Ğ³Ğ¸Ñ‡ĞµÑĞºÑƒÑ Ñ‚Ñ€Ğ°Ğ½ÑĞ»Ğ¸Ñ‚ĞµÑ€Ğ°Ñ†Ğ¸Ñ Ğº ĞµĞ´Ğ¸Ğ½Ğ¾Ğ¼Ñƒ Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ‚Ñƒ
    """
    if pd.isna(text) or not text:
        return ''

    text = str(text)

    # Glottal I â†’ Dotless I (ÑƒĞ½Ğ¸Ñ„Ğ¸ĞºĞ°Ñ†Ğ¸Ñ)
    text = text.replace('ê¼', 'I')  # Ğ—Ğ°Ğ³Ğ»Ğ°Ğ²Ğ½Ğ°Ñ
    text = text.replace('ê½', 'Ä±')  # Ğ¡Ñ‚Ñ€Ğ¾Ñ‡Ğ½Ğ°Ñ â†’ dotless i

    # Ğ£Ğ±Ğ¸Ñ€Ğ°ĞµĞ¼ Ñ‚ĞµÑ…Ğ½Ğ¸Ñ‡ĞµÑĞºĞ¸Ğµ Ğ¼Ğ°Ñ€ĞºĞµÑ€Ñ‹ Ğ¸Ğ· Ñ‚ĞµĞ·Ğ°ÑƒÑ€ÑƒÑĞ°
    text = text.replace('â¸¢', '')
    text = text.replace('â¸£', '')
    text = text.replace('ã€ˆ', '')
    text = text.replace('ã€‰', '')
    text = text.replace('â¸®', '')
    text = text.replace('â‰¡', '')

    # ÑƒĞ±Ğ¸Ñ€Ğ°ĞµĞ¼ combining Ğ´Ğ¸Ğ°ĞºÑ€Ğ¸Ñ‚Ğ¸ĞºĞ¸
    text = text.replace('\u0357', '')

    return text



thesaurus_normalized = [normalize_egyptian_transliteration(line) for line in thesaurus_lemmas]
hf_normalized = [normalize_egyptian_transliteration(line) for line in hf_lemmas]


all_lemmas_normalized = thesaurus_normalized + hf_normalized


output_path = '/content/drive/MyDrive/Ğ”Ñ€ĞµĞ²Ğ½Ğ¸Ğµ ÑĞ·Ñ‹ĞºĞ¸/w2v_data/altes_reich_lemmas_NORMALIZED.txt'
with open(output_path, 'w', encoding='utf-8') as f:
    for line in all_lemmas_normalized:
        if line.strip():
            f.write(line + '\n')

print(f"Ğ˜Ğ¢ĞĞ“ĞĞ’Ğ«Ğ™ ĞĞĞ ĞœĞĞ›Ğ˜Ğ—ĞĞ’ĞĞĞĞ«Ğ™ ĞšĞĞ ĞŸĞ£Ğ¡ ")
print(f"Ğ’ÑĞµĞ³Ğ¾ ÑÑ‚Ñ€Ğ¾Ğº: {len(all_lemmas_normalized)}")

all_chars = set()
for line in all_lemmas_normalized:
    all_chars.update(line)

print(f"\nĞ’ÑĞµĞ³Ğ¾ ÑƒĞ½Ğ¸ĞºĞ°Ğ»ÑŒĞ½Ñ‹Ñ… ÑĞ¸Ğ¼Ğ²Ğ¾Ğ»Ğ¾Ğ²: {len(all_chars)}")

all_words = []
for line in all_lemmas_normalized:
    all_words.extend(line.split())

print(f"\nĞ’ÑĞµĞ³Ğ¾ Ñ‚Ğ¾ĞºĞµĞ½Ğ¾Ğ²: {len(all_words)}")
print(f"Ğ£Ğ½Ğ¸ĞºĞ°Ğ»ÑŒĞ½Ñ‹Ñ… Ğ»ĞµĞ¼Ğ¼: {len(set(all_words))}")

from collections import Counter
freq = Counter(all_words)
print("\nĞ¢Ğ¾Ğ¿-30 Ğ»ĞµĞ¼Ğ¼:")
for lemma, count in freq.most_common(30):
    print(f"  {lemma}: {count}")

print(f"Ğ¡Ğ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ¾ Ğ²: {output_path}")
```
